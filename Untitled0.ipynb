```python
a = 0
b = 0
c = 0
print("nhập mười số nguyên: ")
for i in range(10):
    x = int(input(f"số thứ {i+1}: "))
    if x > 0:
       a += 1
    elif x < 0:
       b += 1
    else:
       c += 1
       print("số dương :", a)
       print(" số âm :" , b)
       print("số bầng 0 :", c)

```

```python
n = int(input("nhập số: "))
if n <= 1:
   print("không phải số nguyên tố")
else:
  la_nt = True
  for i in range(2, int(n**0.5) + 1):
      if n % i == 0:
        la_nt = False
        break
  if la_nt:
    print("là số nguyên tố")
  else:
    print("ko phải số nguyên tố")
```
